```python
raw_train_val = pd.read_csv('/kaggle/input/Kannada-MNIST/train.csv')
raw_test = pd.read_csv('/kaggle/input/Kannada-MNIST/test.csv')
print(raw_train_val.shape)
print(raw_test.shape)
print(raw_train_val.iloc[0:1,0])
```

```python
num = raw_train_val['label'].value_counts()#num is a new Series
sns.barplot(num.index,num)
```

```python
numbers = num.index.values
fig,ax = plt.subplots(2,5,sharex='col',sharey='row',dpi=150)
for i in range(2):
    for j in range(5):
        number_class = raw_train_val[raw_train_val['label']==numbers[i*5+j]]
        number_sample = number_class.iloc[2,:].values[1:].reshape(28,28)
        ax[i,j].set_title(str(numbers[i*5+j]))
        ax[i,j].imshow(number_sample,cmap=plt.get_cmap('Blues'),)
plt.show()    
```
